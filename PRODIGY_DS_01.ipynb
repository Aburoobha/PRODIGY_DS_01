```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_6298256.csv', skiprows=3)
dffinal = df.drop(['Country Code', 'Indicator Name', 'Indicator Code'], axis=1)
dffinal.head()
```

```python
year_column = '2020'

dfsorted = df.sort_values(by=year_column, ascending=False)

plt.figure(figsize=(8, 6))
bars = plt.bar(dfsorted['Country Name'][:5], dfsorted[year_column][:5], label=year_column)

plt.legend()
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

    
plt.title(f'Top 5 Countries by Total Population in {year_column}')
plt.xlabel('Country')
plt.ylabel('Total Population')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
years_1970s = [str(year) for year in range(1970, 1980)]
years_2000s = [str(year) for year in range(2000, 2010)]

dffiltered = df[['Country Name'] + years_1970s + years_2000s]

chosen_country = 'India'

country_data = dffiltered[dffiltered['Country Name'] == chosen_country]

population_1970s = country_data[years_1970s].values.flatten()
population_2000s = country_data[years_2000s].values.flatten()

plt.figure(figsize=(10,6))
bar_width = 0.2

plt.bar(years_1970s, population_1970s, label='1970s', width=bar_width)
plt.bar(years_2000s, population_2000s, label='2000s', width=bar_width)

plt.title(f'Population Growth of {chosen_country} in the 1970s and 2000s')
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.legend()
plt.show()
```
